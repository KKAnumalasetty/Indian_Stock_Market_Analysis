```python
from nsetools import Nse
from nsepy import get_history
import datetime
from datetime import date
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

<b>nsetools</b> python package helps you get real time prices from NSE India stock exchange

<i> Beyond awesome and crazy!!! </i>

 <b> nsepy </b> python package helps you get historical prices from NSE import get_history
>>> from datetime import date
>>> data = get_history(symbol="SBIN", start=date(2015,1,1), end=date(2015,1,31))
>>> data[['Close']].plot()

https://nsetools.readthedocs.io/en/latest/usage.html

```python
# print(datetime.datetime.now())
sbin = get_history(symbol="SBIN", start=date(2000,1,1), end=date.today())
sbin.reset_index(inplace=True)
hdfc = get_history(symbol="HDFC", start=date(2000,1,1), end=date.today())
hdfc.reset_index(inplace=True)
reliance = get_history(symbol="RELIANCE", start=date(2000,1,1), end=date.today())
reliance.reset_index(inplace=True)
infy = get_history(symbol="INFY", start=date(2000,1,1), end=date.today())
infy.reset_index(inplace=True)
# data[['Close']].plot()

# data
fig = go.Figure()
fig.add_trace(go.Scatter(x=sbin["Date"], y=sbin["Close"],
                    mode='lines',
                    name='SBI'))
fig.add_trace(go.Scatter(x=hdfc["Date"], y=hdfc["Close"],
                    mode='lines',
                    name='HDFC'))
fig.add_trace(go.Scatter(x=reliance["Date"], y=reliance["Close"],
                    mode='lines',
                    name='Reliance'))
fig.add_trace(go.Scatter(x=infy["Date"], y=infy["Close"],
                    mode='lines',
                    name='Infosys'))

# sbin_trace = px.line(sbin, x="Date", y="Close", title='SBI')
# sbin_trace = px.line(sbin, x="Date", y="Close", title='SBI')
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
fig.show()
```
